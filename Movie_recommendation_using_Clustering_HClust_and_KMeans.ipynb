```python
# Import required libraries

import numpy as np
import pandas as pd
from scipy.spatial import distance
from scipy.spatial.distance import pdist
from sklearn.cluster import KMeans as kmeans, AgglomerativeClustering as hclust
from scipy.cluster.hierarchy import dendrogram, ward
```

```python
# Load the data

movies=pd.read_csv("http://files.grouplens.org/datasets/movielens/ml-100k/u.item",header=None,sep='|',encoding='iso-8859-1')
movies.head(5)
```

```python
# Columns labels are missing, lets add the names

movies.columns=["ID", "Title", "ReleaseDate", "VideoReleaseDate", "IMDB", "Unknown", "Action", "Adventure", "Animation", "Children", "Comedy", "Crime", "Documentary", "Drama", "Fantasy", "FilmNoir", "Horror", "Musical", "Mystery", "Romance", "SciFi", "Thriller", "War", "Western"]
```

```python
# Check the shape of data

print(f"We have {movies.shape[0]} Movies and {movies.shape[1]} features.")
```

```python
# Check summary of the dataset

movies.info()
```

```python
# ID, ReleaseDate, VideoReleaseDate and IMDB are unnecessary columns. Lets delete them.

movies=movies.drop(["ID","ReleaseDate","VideoReleaseDate","IMDB"],axis=1)
```

```python
movies.head()
```

```python
# Check null values

movies.isnull().sum()
```

```python
# Check for duplicate rows.

movies.duplicated().value_counts()
```

```python
# Looks like we have 18 duplicates, lets remove them.

movies.drop_duplicates(inplace=True)
```

```python
# Calculating euclidean distance between each and every point.

dist=pdist(movies.iloc[:,1:20],'euclidean')
len(dist)
```

```python
# Manual calculation 

1664*1663/2
```

```python
# Modelling Part - HClust with 10 clusters

hcm = hclust(affinity="euclidean",linkage="ward",n_clusters=10)
hcm.fit(movies.iloc[:,1:20])
```

```python
# Check the cluster labels

print (f"Number of Clusters : {hcm.n_clusters} \nLabels : {hcm.labels_} \nDatapoints : {len(hcm.labels_)}")
```

```python
# Unique lables and number of datapoints against that label

np.unique(hcm.labels_,return_counts=True)
```

```python
# linkage matrix Z

Z=ward(movies.iloc[:,1:20],)
Z
```

```python
# Plot the hierarchical clustering as a dendrogram

import matplotlib.pyplot as plt 
plt.figure(figsize=(8,8))
x = dendrogram(Z)
```

```python
# Add a new column in cluster in movies dataset which shows cluster label for each movie

movies['hclust_label'] = hcm.labels_
```

```python
movies.head()
```

```python
# Interpret hclust_labels

movies.groupby(by='hclust_label').mean()

# Cluster no 0 contains movies with comedy genre
# Cluster 1 contains movies with horror and documentary genre
```

```python
# Display ten movies that belong to cluster no 1

movies[movies.hclust_label==1].head(10)
```

```python
# Modelling with Kmeans Clustering

kmc = kmeans(n_clusters=10,random_state=42)
kmc.fit(movies.iloc[:,1:20])
```

```python
# Check the lables

kmc.labels_
```

```python
# Check unique labels and number of datapoints

np.unique(kmc.labels_,return_counts=True)
```

```python
# Add a new column in cluster in movies dataset which shows K-means cluster label for each movie

movies['kmeansclust_labels'] = kmc.labels_
movies.head()
```

```python
# Interpret kmeansclust_labels

movies.groupby('kmeansclust_labels').mean()

# Cluster no 0 contains movies with comedy genre
# Cluster 1 contains movies with horror and documentary genre
```

```python
"""
Now if we want to find movies related to a particular movie - 
Step1: find the cluster of that movie
Step2: find all movies in that cluster
"""


# we want to find movies similar to "toy story"
movies[movies.Title.str.match("toy story",case=False)]

# This movie belongs to hclust_label 6 and kmeansclust_labels 4
# we can use any of them
```

```python
# Display ten movies that belong to hclust_label no 6

movies[movies["hclust_label"]==6].head(10)
```

```python
# Display ten movies that belong to kmeansclust_labels no 4

movies[movies["kmeansclust_labels"]==4].head(10)
```
